```python
#import library packages
import pandas as p
import matplotlib.pyplot as plt
import seaborn as s
import numpy as n
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data = p.read_csv("crop.csv")
```

```python
df = data.dropna()
```

```python
df.columns
```

```python
df['rainfall'].hist(figsize=(8,5), color='purple', alpha=0.7)
plt.xlabel('Rainfall Range')
plt.ylabel('Total Count')
plt.title('Rainfall Range with Count')
```

```python
df['ph'].hist(figsize=(8,5), color='c', alpha=0.7)
plt.xlabel('PH Range')
plt.ylabel('Count')
plt.title('PH values with Count')
```

```python
#Propagation by variable
def PropByVar(df, variable):
    dataframe_pie = df[variable].value_counts()
    ax = dataframe_pie.plot.pie(figsize=(8,8), autopct='%1.2f%%', fontsize = 10)
    ax.set_title(variable + ' \n', fontsize = 15)
    return n.round(dataframe_pie/df.shape[0]*100,2)
PropByVar(df, 'label')
```

```python
# Heatmap plot diagram
fig, ax = plt.subplots(figsize=(13,6))
s.heatmap(df.corr(), ax=ax, annot=True)
```
